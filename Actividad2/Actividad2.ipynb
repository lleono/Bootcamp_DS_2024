Introduccion: Se pretende analizar el promedio de muertes en colombia de forma accidental en elos años 2016 al 2019

Pregunta problema (objetivo) en que meses del año hubo mas muestres accidentales de mujeres en los años 2016 al 2019 

Fuente de datos: Datos abiertos (https://www.datos.gov.co/Justicia-y-Derecho/Muertes-accidentales-Colombia-a-os-2016-a-2019/irg9-q5fx/about_data)

EDA (Exploratory Data Analysis)

```python
C1 =T[['Sexo de la victima','Grupo de edad de la victima','Mes del hecho' ]]
sns.boxenplot(data=C1, x= 'Sexo de la victima', y='Mes del hecho', showfliers=False)
plt.xlabel('Sexo')
plt.ylabel('Mes')
plt.title ('Boxplot de genero')
```

```python
C1 =T[['Sexo de la victima','Grupo de edad de la victima','Mes del hecho' ]]
sns.boxplot(data=C1, x= 'Sexo de la victima', y='Mes del hecho', hue='Grupo de edad de la victima', showfliers=False)
plt.xlabel('Sexo')
plt.ylabel('Mes')
plt.title ('Boxplot de Edad')
```

El reporte de muertes en hombres y mujeres tienen tendencias muy parecidas

```python
sns.boxplot(data=C1, x='Sexo de la victima', y='Mes del hecho',showfliers= True)
```
